```python
import pandas as pd
import matplotlib.pyplot as plt

file = "tempdens5503.dat"

data = pd.read_csv(file, sep = "\s+",
                   usecols = [0,1,2],
                   names = ["time", "temp", "dens"])
plt.figure(figsize = (12,4),facecolor = "w")
plt.subplot(1,2,1)
plt.plot(data["time"], data["temp"])
plt.yscale("symlog")
plt.xlim(0,1.5)
plt.ylim(1.0e8, 1.0e10)
plt.subplot(1,2,2)
plt.plot(data["time"], data["dens"])
plt.yscale("symlog")
plt.xlim(0,1.5)
plt.ylim(1.0e5, 1.0e10)
plt.show()
```
