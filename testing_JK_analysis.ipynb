```python
import numpy as np
import pandas as pd
import scipy.spatial
import math as math
import glob

from astropy.stats import jackknife_resampling
from astropy.stats import jackknife_stats
files = glob.glob("../std_def_final/part_IGEs_SCD_PD_Z_*")

def function_1(array):
    
    arr = array.to_numpy()
    resampled_data = jackknife_resampling(arr)
    estimate, bias, error, conf_internval = jackknife_stats(arr, np.mean, 0.95)

    return estimate, bias, error, conf_internval

for file in files:
    print(file)
    file = pd.read_csv(file, 
                   sep = "\s+", usecols = [0, 1, 2, 3, 4], 
                   names = ["p_tag", "Fe", "Ni", "Mn", "Cr"])



    Fe_estimate, Fe_bias, Fe_error, Fe_conf_internval = function_1(file["Fe"])
    Ni_estimate, Ni_bias, Ni_error, Ni_conf_internval = function_1(file["Ni"])
    Mn_estimate, Mn_bias, Mn_error, Mn_conf_internval = function_1(file["Mn"])
    Cr_estimate, Cr_bias, Cr_error, Cr_conf_internval = function_1(file["Cr"])


    Ni_Fe_estimate = Ni_estimate/Fe_estimate 
    Mn_Fe_estimate = Mn_estimate/Fe_estimate
    Cr_Fe_estimate = Cr_estimate/Fe_estimate 


    Ni_Fe_error = Ni_error/Fe_error 
    Mn_Fe_error = Mn_error/Fe_error 
    Cr_Fe_error = Cr_error/Fe_error 


    Ni_Fe_conf_internval = Ni_conf_internval / Fe_conf_internval #[Ni_conf_internval[0]/Fe_conf_internval[0], Ni_conf_internval[1]/Fe_conf_internval[1]]
    Mn_Fe_conf_internval = Mn_conf_internval / Fe_conf_internval #[Mn_conf_internval[0]/Fe_conf_internval[0], Mn_conf_internval[1]/Fe_conf_internval[1]]
    Cr_Fe_conf_internval = Cr_conf_internval / Fe_conf_internval #[Cr_conf_internval[0]/Fe_conf_internval[0], Cr_conf_internval[1]/Fe_conf_internval[1]]
    
    error_bars = [(Ni_Fe_conf_internval[1] - Ni_Fe_conf_internval[0])/2, 
                  (Mn_Fe_conf_internval[1] - Mn_Fe_conf_internval[0])/2,
                  (Cr_Fe_conf_internval[1] - Cr_Fe_conf_internval[0])/2]
    
    #print(sum(file["Ni"]/sum(file["Fe"])))
    print("Estiamtes :", Ni_Fe_estimate, Mn_Fe_estimate, Cr_Fe_estimate)
    print("Errors : ", Ni_Fe_error, Mn_Fe_error, Cr_Fe_error)
    print("Conf. Interval :" , Ni_Fe_conf_internval,  Mn_Fe_conf_internval, Cr_Fe_conf_internval)
    print("Conf. Interval lower :", "%.5e" %Ni_Fe_conf_internval[0], "%.5e" %Mn_Fe_conf_internval[0], "%.5e" %Cr_Fe_conf_internval[0])
    print("Conf. Interval upper:", "%.5e" %Ni_Fe_conf_internval[1], "%.5e" %Mn_Fe_conf_internval[1], "%.5e" %Cr_Fe_conf_internval[1])
    print("error bars top", (Ni_Fe_conf_internval[1] - Ni_Fe_estimate), (Mn_Fe_conf_internval[1] - Mn_Fe_estimate), (Cr_Fe_conf_internval[1] - Cr_Fe_estimate))
    print("error bars bottom", (Ni_Fe_conf_internval[0] - Ni_Fe_estimate), (Mn_Fe_conf_internval[0] - Mn_Fe_estimate), (Cr_Fe_conf_internval[0] - Cr_Fe_estimate))
    
    print("")
    print("")
```
