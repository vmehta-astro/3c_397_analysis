```python
import numpy as np
import pandas as pd
import glob 

files = glob.glob("./SCD_PD_*")

print(files)

data_0 = pd.read_csv(files[0], sep = '\s+', usecols = [0,1], names = ['species_names', 'mass_frac'])
data_0.drop(index = data_0.index[0], axis = 0, inplace=True)
#data_0 = data_0[data_0['mass_frac'] >=1.0e-6]
data_0.reset_index(drop = True, inplace = True)
#print(data_0)


data_1 = pd.read_csv(files[1], sep = '\s+', usecols = [0,1], names = ['species_names', 'mass_frac'])
data_1.drop(index = data_1.index[0], axis = 0, inplace=True)
#data_1 = data_1[data_1['mass_frac'] >=1.0e-6]
data_1.reset_index(drop = True, inplace = True)
#print(data_0)

data_2 = pd.read_csv(files[2], sep = '\s+', usecols = [0,1], names = ['species_names', 'mass_frac'])
data_2.drop(index = data_1.index[0], axis = 0, inplace=True)
#data_2 = data_2[data_2['mass_frac'] >=1.0e-6]
data_2.reset_index(drop = True, inplace = True)
#print(data_2)

    
pub_data = pd.DataFrame()
pub_data['species_names'] = data_0['species_names'].values
pub_data['mass_frac_z_0'] = data_2['mass_frac'].values
pub_data['mass_frac_z_1p5'] = data_1['mass_frac'].values
pub_data['mass_frac_z_3'] = data_0['mass_frac'].values

pub_data = pub_data.drop(pub_data[(pub_data['mass_frac_z_0'] <= 1.0e-6) & (pub_data['mass_frac_z_1p5'] <= 1.0e-6) & (pub_data['mass_frac_z_3'] <= 1.0e-6)].index)

#data = pub_data[['mass_frac_z_0','mass_frac_z_1p5','mass_frac_z_3']].apply(lambda x: np.where(x < 1.0e-6, '-', x))
#pub_data['mass_frac_z_0'], pub_data['mass_frac_z_1p5'], pub_data['mass_frac_z_3'] = data['mass_frac_z_0'], data['mass_frac_z_1p5'], data['mass_frac_z_3']
pub_data.reset_index(drop = True, inplace = True)
print(pub_data)

#print(data_0)
    
    

```
