```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.spatial


mass = 2.78258e33
particles = 9990

mass_per_particle = mass/particles


data = pd.read_csv('./particles.dat', sep = '\s+', 
                   usecols = [0,1,2,3,4], 
                   names = ['X_Fe', 'X_Cr', 'X_Cr_to_X_Fe','X_pos','Y_pos'])

data.sort_values(by = ['X_Cr_to_X_Fe'], ascending = False, inplace = True)

ejecta_particles = len(data)
ejecta_mass = ejecta_particles * mass_per_particle



z = (data['X_Cr']/data['X_Fe'])
x = data['X_pos']
y = data['Y_pos']
fig = plt.figure()

grid = AxesGrid(fig,rect = (0.075, 0.075, 5, 3),
                nrows_ncols = (3, 5),
                axes_pad= 2.0,
                label_mode="2",
                share_all = True,
                cbar_location = "right",
                cbar_mode = "single",
                cbar_size = "6%",
                cbar_pad = "10%")

for slc, j in slc_arr:
    p = z[j]
    q = z[z == p].index.tolist()
    x0 = x[q].to_numpy()
    y0 = y[q].to_numpy()

    points = list(zip(x,y))
    ckdtree = scipy.spatial.cKDTree(points)
    
    for i in range(len(x0)):
        
        X = x0[i]
        Y = y0[i]
        clump_points = []
        clumps_data = []
        
        for increment in range(1,200):
            clump_points = ckdtree.query([X,Y],increment)
            clumps_data = data.iloc[clump_points[1]]
            
            if (((clumps_data["X_Fe"]).sum())/(data["X_Fe"].sum())) >= 0.06:
                break
        
        print(clump_points[1])
        print(len(clumps_data))
        print(((clumps_data["X_Fe"]).sum())/(data["X_Fe"].sum()))
        print(clumps_data["X_Cr"].sum() / clumps_data["X_Fe"].sum())  

    cl_z = clumps_data["X_Cr"] / clumps_data["X_Fe"]
    
    plot = slc.plots[("gas","density")]
    plot.figure = fig
    plot.axes = grid[j].axes
    plot.cax = grid.cbar_axes[j]
    slc.render()
    axes_locations = grid[j].get_axes_locator()
    
    scatter_plot = grid[j].inset_axes([0.0, 0.0, 1.0, 1.0], facecolor ='None')
    scatter_plot.scatter(clumps_data["X_pos"], clumps_data["Y_pos"], s = (60 * ((cl_z - cl_z.min())/cl_z.max() - cl_z.min())) + 0.5 , c = cl_z, cmap = 'Reds')
    scatter_plot.set_xlim(0,2*cent*1.0e-5)
    scatter_plot.set_ylim(-2*cent*1.0e-5,2*cent*1.0e-5)
    scatter_plot.tick_params(labelsize = 16)
    yfmt = ScalarFormatterForceFormat()
    yfmt.set_powerlimits((0,0))
    scatter_plot.yaxis.set_major_formatter(yfmt)
    scatter_plot.xaxis.set_major_formatter(yfmt)
    scatter_plot.ticklabel_format(axis = 'both', style = 'sci', useMathText= True, scilimits = [0,0])
    scatter_plot.tick_params(axis='both',labelsize = 20)
    scatter_plot.xaxis.get_offset_text().set_fontsize(20)
    scatter_plot.yaxis.get_offset_text().set_fontsize(20)
    scatter_plot.set_xlabel('r (km)', fontsize = 20)
    scatter_plot.set_ylabel('z (km)', fontsize = 20)
    #s = (60 * ((z - z.min())/z.max() - z.min())) + 0.5
    
#fig.set_tight_layout(True)
plt.savefig("test_plots.png")
#slc.render()
plt.show()
```
