```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('../HIGH_DEN_DEF_8_sec_final/time_ratio.dat', sep = '\s+',
                 usecols = [0,1], names = ['time', 'ratio']) 
print(df.to_string())
```

```python
fig, ax = plt.subplots()
ax.plot(df['time'], df['ratio'], ls = '-', color = 'red', marker = 'o')
#ax.set_yscale('log')
ax.set_title(r'$M_{ejecta}$/$M_{total}$ Plot')
ax.set_ylabel(r'$M_{ejecta}$/$M_{total}$')
ax.set_xlabel('Time (s)')
plt.show()
```

```python
tf = float(df['ratio'].iloc[-1])

i_1 = df.index[round(df['time'],2) == 7.00].tolist()
tf_1 = float(df['ratio'].iloc[i_1])
r1 = ((tf - tf_1)/tf_1)*100

i_2 = df.index[round(df['time'],2) == 6.00].tolist()
tf_2 = float(df['ratio'].iloc[i_2])
r2 = ((tf - tf_2)/tf_2)*100

i_3 = df.index[round(df['time'],2) == 5.00].tolist()
tf_3 = float(df['ratio'].iloc[i_3])
r3 = ((tf - tf_3)/tf_3)*100

print('increase in the ratio(M_eje/M_tot) between t = 5s and t = 6s :', r1, '%')
print('increase in the ratio(M_eje/M_tot) between t = 4s and t = 6s :', r2, '%')
print('increase in the ratio(M_eje/M_tot) between t = 3s and t = 6s :', r3, '%')
```
