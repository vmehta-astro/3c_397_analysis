```python
import numpy as np
import pandas as pd
import glob
import h5py
import matplotlib.pyplot as plt

files = glob.glob("def_high_dens_hdf5_part_044347")
files.sort()



for file in files:
    f = h5py.File(file ,"r")
    dset = list(f.keys())
    print(dset)
    
    
    r_s = f['real scalars']
#    print(r_s[...])
    i_r_p = f['particle names']
#    print(i_r_p[...])
#    print(len(i_r_p))
    df = pd.DataFrame(data = i_r_p, dtype = str)
    
    t_p = f['tracer particles']
    data = pd.DataFrame(data = t_p)# columns = r_s)
#data.sort_values(by = [4], ascending = False, inplace = True)

#data = data[data[7] <= 2.16e7]  
#data = data[data[7] >= 1.84e7]
#data = data[data[6] <= 4.5e6]

data.sort_values(by = [13], ascending = True, inplace = True)
data.reset_index(drop = True, inplace = True)
data
```

```python
density = data[:][1]
pres = data[:][11]
internal_e = data[:][2]
vel_x = data[:][15]
vel_y = data[:][16]
grav_pot = data[:][4]
part_tag = data[:][13]


vel_magnitude_square = np.square(vel_x) + np.square(vel_y)
```

```python
plt.figure(figsize = (10,6))
plt.scatter(part_tag, 0.5 * vel_magnitude_square, color = "red", s = 3, label = "Kinetic Energy")
plt.scatter(part_tag, internal_e, color = "blue", s = 3, label = "Internal Energy")
plt.scatter(part_tag, (grav_pot), color = "green", s = 3, label = "Grav. Potential")
#plt.scatter(vel_magnitude_square, part_tag ,color = "red")
plt.legend(loc = 'lower left')
plt.yscale("symlog")
#plt.ylim(1.0e12,1.0e19)
plt.show()
```
