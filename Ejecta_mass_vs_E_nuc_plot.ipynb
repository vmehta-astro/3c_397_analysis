```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#file = "Ejecta_mass_and_E_nuc_data.dat"

#df = pd.read_csv(file, sep = "\s+", usecols = [0], names = ["Ejecta_mass"])

Total_E_nuc_high_dens = 4.869905801403155e+50
Ejecta_mass_high_dens = 2.7772999380317092e+32 / 1.989e33
Cent_dens_high_dens = 6.0e9

Total_E_nuc_std_dens = 5.82362382256548e+50
Ejecta_mass_std_dens = 9.604583059276018e+32 / 1.989e33
Cent_dens_std_dens = 2.2e9


Total_E_nuc_Jordan = 0.89 * 4.5e50
Ejecta_mass_Jordan = 0.13 + 0.03 + 0.07
Cent_dens_Jordan = 2.0e9

Total_E_nuc_Kromer = 4.9e50
Ejecta_mass_Kromer = 0.37 
Cent_dens_Kromer = 2.9e9



#Total_E_nuc = df["E_nuc"].sum()
#print(Total_E_nuc)
plt.figure(figsize = (9,6), facecolor = "w")
plt.title("Ejecta Mass vs Nuclear Energy")

plt.plot(Total_E_nuc_high_dens, Ejecta_mass_high_dens, "ob", label = "High Density model" + r' ($\rho_c = 6.0*10^9$)')
plt.plot(Total_E_nuc_std_dens, Ejecta_mass_std_dens, "or", label = "Std. Density model" + r' ($\rho_c = 2.2*10^9$)')

plt.plot(Total_E_nuc_Jordan, Ejecta_mass_Jordan, "Dg", label = "Jordan et al 2D model" + r' ($\rho_c = 2.0*10^9$)')
plt.plot(Total_E_nuc_Kromer, Ejecta_mass_Kromer, "Dm", label = "Kormer et al 3D model" + r' ($\rho_c = 2.9*10^9$)')


#plt.xscale("symlog")
plt.xlim(1.0e50, 6.0e50)
#plt.ylim(1.0e32, 1.0e33)
plt.xlabel("Nuclear Energy (erg)")
plt.ylabel("Ejecta Mass" + r' ($M_\odot$)')
plt.legend(loc = "upper left")
plt.savefig("Ejecta_mass_VS_E_nuc_plot.png")

plt.figure (figsize = (9,6), facecolor = "w")
plt.title("Central Density vs Nuclear Energy")

plt.plot(Total_E_nuc_high_dens, Cent_dens_high_dens,"ob", label = "High Density model")
plt.plot(Total_E_nuc_std_dens, Cent_dens_std_dens, "or", label = "Std. Density model")
plt.plot(Total_E_nuc_Jordan, Cent_dens_Jordan, "Dg", label = "Jordan et al 2D model")
plt.plot(Total_E_nuc_Kromer, Cent_dens_Kromer, "Dm", label = "Kormer et al 3D mode")

plt.xlim(1.0e50, 6.0e50)
plt.xlabel("Nuclear Energy (erg)")
plt.ylabel("Central Density" + r' $(gm/cm^3)$')
plt.legend(loc = "upper left")
plt.savefig("Central_density_VS_E_nuc_plot.png")
```

```python
import math 

high_rho = 6.0e9
log_high_rho = math.log10(high_rho)
print(log_high_rho)

log_high_rho_ash = 9.71
high_rho_ash = 10**(log_high_rho_ash)
print("%4E" %high_rho_ash)

A_high = (high_rho - high_rho_ash)/(high_rho_ash + high_rho)
print(A_high)
```

```python
fl_speed_high = 124.4

high_density_model_radius = 1600

r_crit_high = 19 * (fl_speed_high / 100 ) * ((0.09/A_high)**(0.5)) * ((2.2e9/high_rho)**(0.5))

print(r_crit_high)

```
