```python
import numpy as np
import matplotlib.pyplot as plt

High_dens_pure_def = np.array([9.060e+06, 1.374e+07, 1.130e+06, 3.131e+07, 2.445e+07])
High_dens_DDT = np.array([3.513e+07, 7.194e+07, 1.536e+07, 2.941e+06, 1.704e+05])

species = np.array(['Fe', 'Ni', 'Mn', 'Cr', 'Ti'])
X = np.arange(5)
fig = plt.figure(figsize = (10,6), facecolor = 'w')
plt.bar(X, High_dens_pure_def, width = 0.20, color = 'g', label = 'HIGH-DENS PURE-DEF')
plt.bar(X + 0.25, High_dens_DDT, width = 0.20, color = 'b', label = 'HIGH-DENS DDT')
plt.yscale('log')
plt.xlabel('Abundances', size = 15)
plt.ylabel(r'$Z\ (Z_{\odot})$', size = 15)
plt.ylim(1.0e5, 1.0e9)
plt.xticks(X+0.25/2, species, size = 12)
plt.yticks(size = 12)
plt.legend(loc = 'upper right')
plt.show()
fig.savefig('plot_with_mass_in_sol.png')
print(High_dens_pure_def/High_dens_DDT)
```
