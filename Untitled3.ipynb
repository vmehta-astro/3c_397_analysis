```python
import yt
import glob

files = glob.glob("./def_high_dens_hdf5_plt_cnt_000088")

cent = 1.25e8

for file in files:
    ds = yt.load(file)
    slc = yt.SlicePlot(ds, 'theta', [("gas", "temperature")], origin = "native", center = [cent,0,0])
    slc.set_width((2*cent,4*cent))
    slc.set_cmap(field = "temperature", cmap = "inferno")
    slc.set_zlim("temperature", 1.0e7,1.0e10)
    slc.show()
```
