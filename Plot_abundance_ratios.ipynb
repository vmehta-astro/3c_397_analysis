```python
import numpy as np
import matplotlib.pyplot as plt

font = {'family': 'serif',
        'color':  'black',
        'weight': 'normal',
        'size': 20,
        }

#High-dens Pure Deflagration Model
High_dens_pure_def = np.array([(3.12708e-01, 3.21192e-01, 3.29716e-01),      #Ni/Fe
                               (3.03566e-02, 3.37995e-02, 3.62006e-02),      #Mn/Fe
                               (7.28456e-02, 7.26775e-02, 7.26309e-02),      #Cr/Fe
                               (1.39597e-02, 1.40522e-02, 1.41135e-02)])     #Ti/Fe

#High-dens DDT Model    
High_dens_DDT = np.array([(3.61149e-01, 3.96093e-01, 4.32417e-01),      #Ni/Fe
                          (2.12103e-02, 2.29942e-02, 2.48535e-02),      #Mn/Fe
                          (2.88080e-03, 2.96677e-03, 3.10682e-03),      #Cr/Fe
                          (3.52808e-05, 2.95869e-05, 2.52204e-05)])     #Ti/Fe

#Std-dens Pure Deflagration Model
Std_dens_pure_def = np.array([(1.35503e-01, 1.52476e-01, 1.67356e-01),     #Ni/Fe
                              (1.34329e-02, 2.11424e-02, 2.68982e-02),     #Mn/Fe
                              (1.55698e-02, 1.45406e-02, 1.46152e-02),     #Cr/Fe
                              (5.72986e-04, 5.50388e-04, 5.49063e-04)])    #Ti/Fe

Z = np.array([(0,1.5,3.0),(0,1.5,3.0),(0,1.5,3.0),(0,1.5,3.0)])

fig, ax = plt.subplots(1,3, figsize = (25,10), facecolor = 'w')
markers = np.array(['o','s','>'])
y_axis = np.array([r'$\mathrm{Ni/Fe}$', r'$\mathrm{Mn/Fe}$', r'$\mathrm{Cr/Fe}$', r'$\mathrm{Ti/Fe}$'])
y_axis_lims = np.array([(4.0e-2,5.0e-1), (3.0e-4,4.0e-1), (7.0e-4,5.0e-1), (1.0e-5, 5.0e-1)])

fill_space = np.linspace(-1,4, 3)
upper_bound = np.array([0.24, 0.033, 0.034, 0.018])
lower_bound = np.array([0.12, 0.018, 0.021, 0.01])
observed_values = np.array([(0.17,0.17,0.17),
                            (0.025, 0.025, 0.025),
                            (0.027, 0.027, 0.027)])

#error bars

Y_err_HCD_PD = 1.96 * np.array([(),
                                (),
                                ()])
Y_err_SCD_PD = 1.96 * np.array([(),
                                (),
                                ()])

#southern_upper_bound = np.array([0.25, 0.060, 0.117, 0.018])
#southern_lower_bound = np.array([0.15, 0.042, 0.097, 0.010])

#western_upper_bound = np.array([0.32, 0.056, 0.038, 0.002])
#western_lower_bound = np.array([0.24, 0.044, 0.030, 1.0e-5])

for i in range(len(High_dens_pure_def) - 1):
    
    if i < 3:
        ax[i].plot(fill_space, observed_values[i], color = 'black',linewidth = '3',linestyle = 'dashed',label = "Observed 3C 397 ratios")
        ax[i].fill_between(fill_space, upper_bound[i], lower_bound[i], alpha = 0.5, color = 'black', label = "Observational Error bars")
    #ax[i].fill_between(fill_space, southern_upper_bound[i], southern_lower_bound[i], alpha = 0.3, color = 'blue',hatch = 'O' ,label = "Observed: 3C 397 Southern region")    
    #ax[i].fill_between(fill_space, western_upper_bound[i], western_lower_bound[i], alpha = 0.3, color = 'red',hatch = 'X' ,label = "Observed: 3C 397 Western region")
    
    
    ax[i].plot(Z[i], High_dens_pure_def[i], color = 'blue', marker = 'o', linewidth = '2',markersize = 15,linestyle = 'dashed', label = "DEF-HIGHDENS-LOWC/NR-CENTER")
    #ax[i].errorbar(Z[i], High_dens_pure_def[i], yerr = 10 * Y_err_HCD_PD[i], fmt = 'o', markersize = 10, capsize = 14)
    
    ax[i].plot(Z[i], High_dens_DDT[i], color = 'red', marker = 's', linewidth = '2', markersize = 15,linestyle = 'dashed', label = "DDT-HIGHDENS-LOWC/HIGHOFF")
    
    
    ax[i].plot(Z[i], Std_dens_pure_def[i], color = 'green', marker = '>', linewidth = '2', markersize = 15,linestyle = 'dashed',label = "DEF-STDDENS/NR-CENTER")

    
    ax[i].set_xlabel(r"$\mathrm{Z\ (Z_{\odot})}$", fontsize = 28, fontdict = font)
    ax[i].set_ylabel(y_axis[i], fontsize = 28, fontdict = font)
    ax[i].set_xlim(-0.1,3.1)
    ax[i].set_ylim(y_axis_lims[i])
    ax[i].set_yscale('log')
    ax[i].tick_params(labelsize = 24, length = 20, width = 2, which='major', direction = 'in')
    ax[i].tick_params(length = 12, width = 1, which = 'minor', direction = 'in')
    ax[i].xaxis.set_ticks_position('both')
    ax[i].yaxis.set_ticks_position('both')
    if i == 0:
        ax[i].legend(loc = 'lower right', prop={'size': 15})
    fig.tight_layout(pad=5.0)
plt.show()
fig.savefig("abundance_ratios_plot.png")
```
